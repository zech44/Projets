```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#1
df = sns.load_dataset('taxis')
df.head(7)
```

```python
#2
df.rename(columns={"pickup": "depart"}, inplace = True)
df.rename(columns={"dropoff" : "arrive"}, inplace = True)
df.rename(columns={"passengers" : "nbpassagers"}, inplace = True)
df.rename(columns={"fare" : "tarif"}, inplace = True)
df.rename(columns={"tip" :"pourboire"}, inplace = True)
df.rename(columns={"tolls" :"peage"}, inplace = True)
df.rename(columns={"color" :"couleur"}, inplace = True)
df.rename(columns={"pickup_zone" :"zone_depart"}, inplace = True)
df.rename(columns={"dropoff_zonz" :"zone_arrivee"}, inplace = True)
df.rename(columns={"pickup_borough" :"arrondissement_depart"}, inplace = True)
df.rename(columns={"dropoff_borough" :"arrondissement_arrivee"}, inplace = True)
df.head(7)
```

```python
#11
tableau_croise = pd.crosstab(df['nbpassagers'], df['payment'])
print(tableau_croise)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(tableau_croise, annot=True, fmt="d", cmap="YlGnBu", cbar=True, linewidths=.5)

# Ajouter des labels pour les axes
plt.xlabel('Type de paiement')
plt.ylabel('Nombre de passagers')

# Ajouter un titre Ã  la heatmap
plt.title('Heatmap du nombre de passagers par type de paiement')

# Afficher la heatmap
plt.show()
```
