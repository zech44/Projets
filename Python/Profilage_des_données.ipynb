```python
# Calcul de la moyenne d'un échantillon
sample=[1,2,3,4,4,5,6]
moyenne=sum(sample)/len(sample)
print(moyenne)
```

```python
 import statistics
# Calcul de la variance 
variance=statistics.variance(sample)
print(variance)
# Calcul de la médiane 
mediane=statistics.median(sample)
print(mediane)
# le mode
mode=statistics.mode(sample)
print(mode)
# l'écart type 
ecart_type=statistics.stdev(sample)
print(ecart_type)
```

```python
# Simulation d'un lancer de dé
# Tirage sans remise de 5 valeurs
import random
de=[1,2,3,4,5,6]
tirage=random.sample(de,5)
print(tirage)

# Tirage avec remise 
tirage2=random.choices(de,k=5)
print(tirage2)
```

```python
# Le théorème centrale limite (TCL)
nb_echan=1000
taille=1
sample=np.random.randint(1,7,(nb_echan,taille))
moyenne_echan=sample.mean(axis=1)
```

```python
sns.histplot(moyenne_echan, stat="density",kde=True, bins=6)
```

```python
# Lanceur de plusieurs dés 
plt.figure(figsize=(12,8))
nb_echan=10000
taille=[1,2,5,100,200]
for i,j in enumerate(taille,1) :
    sample=np.random.randint(1,7,(nb_echan,j))
    moyenne_echan=sample.mean(axis=1)
    plt.subplot(2,3,i)
    sns.histplot(moyenne_echan, stat="density", kde=True)
```
