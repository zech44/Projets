```python
import pandas as pd
import numpy as np
from scipy.stats import kurtosis, skew, jarque_bera
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_arch
```

```python
df = pd.DataFrame([[3.2,4.1,4.5,5,6.2,4.8,3.5,4,2.8,1.7,4.2,5.1,4.9,3.4,2.6,1.6,0.2,1,2.5,2.8,-2.1,1.3,3,2.5,2.6],
                [2.5,3,2.8,1.9,2.3,3.1,2.5,1.8,5.4,3,1.5,2.7,3,2.1,1.9,2.7,1.5,0.9,1.6,2.3,2.8,1.6,5.6,5.6,4.5]], index = ["taux_C","taux_I"])

df
```

```python
cemac = df.T
cemac
```

```python
resume = cemac.describe().T
resume
```

```python
resume["skewness"] = skew(cemac)
resume["kurtosis"] = kurtosis(cemac)
resume
```

```python
jb_stat, jb_pvalue = jarque_bera(cemac["taux_C"])
jb_stat1, jb_pvalue1 = jarque_bera(cemac["taux_I"])
x, y, z, t = jb_stat, jb_pvalue, jb_stat1, jb_pvalue1
resume["stat"] = [x, z]
resume["pvalue"] = [y, t]
```
