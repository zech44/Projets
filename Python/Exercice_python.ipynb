```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Charger le dataset Titanic
df = pd.read_csv('titanic3.csv',sep=";")
```

```python
df.head()
```

```python
#Infos sur le dataset : types de variables, valeurs manquantes, etc.
df.info()
```

```python
#Statistiques descriptives
df.describe()
```

```python
# Vérifier les valeurs manquantes
df.isnull().sum()
```

```python
# Pourcentage de passagers ayant survécu
survived_percentage = df['survived'].mean() * 100
print(f"Pourcentage de passagers ayant survécu : {survived_percentage:.2f}%")
```

```python
# Diagramme à barres du nombre de survivants et non-survivants (avec countplot de seaborn)
sns.countplot(x='survived', data=df)
plt.title('Nombre de survivants et non-survivants')
plt.show()
```

```python
# Comparaison du taux de survie par sexe avec countplot de seaborn)
sns.countplot(x='survived', hue='sex', data=df)
plt.title('Taux de survie par sexe')
plt.show()
```

```python
# Taux de survie par sexe
survival_by_sex = df.groupby('sex')['survived'].mean() * 100
print(survival_by_sex)
```

```python
# Visualiser la survie par classe (avec countplot de seaborn)
sns.countplot(x='survived', hue='pclass', data=df)
plt.title('Survie par classe')
plt.show()
```

```python
# Taux de survie par classe
survival_by_class = df.groupby('pclass')['survived'].mean() * 100
print(survival_by_class)
```

```python
# Affichage de la distribution des âges (avec histplot de seaborn)
sns.histplot(df['age'].dropna(), kde=True)
plt.title('Distribution des âges')
plt.show()
```

```python
# Boîte à moustache de l'âge et survie (Boxplot de seaborn)
sns.boxplot(x='survived', y='age', data=df)
plt.title('Âge et survie')
plt.show()
```

```python
# Statistiques d'âge des survivants vs non-survivants
age_stats = df.groupby('survived')['age'].describe()
print(age_stats)
```

```python
# Distribution du prix du billet (avec histplot)
sns.histplot(df['fare'], bins=10)
plt.title("Distribution du prix des billets")
plt.show()
```

```python
#Boîte à moustache du prix en fonction de la classe (Boxplot)
sns.boxplot(x='pclass', y='fare', data=df)
plt.title('Prix du billet par classe')
plt.show()
```

```python
# Diagramme en nuage de points des âges en fonction de la classe et de la survie (scatterplot de seaborn)
sns.scatterplot(x='age', y='fare', hue='survived', style='pclass', data=df)
plt.title("Âge et classe par rapport à la survie")
plt.show()
```

```python
# Remplir les âges manquants avec la médiane
df['age'].fillna(df['age'].median(), inplace=True)
print(df['age'].isnull().sum())
```
