<h4 id="18">18. Quels sont les meilleurs √©tudiants selon le type de maison ?<a href="#sommaire">üëÜ</a></h4>

```python
# Cr√©ons d'une nouvelle colonne 'Appr√©ciation' pour les notes

def attribuer_appreciation(moyenne):
    if moyenne == 0 or moyenne ==1:
        return "Nul"
    elif moyenne <= 8:
        return "Insuffisant"
    elif moyenne <= 10:
        return "Passable"
    elif moyenne <= 12:
        return "Assez bien"
    elif moyenne <= 14:
        return "Bien"
    elif moyenne <= 18:
        return "Tr√®s bien"

df['Appr√©ciation'] = df['Moyenne'].apply(attribuer_appreciation)
df.head()
```

```python
# Tableau crois√© pour calculer la distribution des appr√©ciations par type de maison
# Avec pour param√®tre margins=False
pd.crosstab(df['Maison'], df['Appr√©ciation'], margins=False) 
```

```python
# Cr√©er une heatmap pour visualiser la relation entre Le type de Maison et l'Appr√©ciation
# Avec cmap="YlGnBu", annot=True, fmt="d" comme param√®tres
plt.figure(figsize=(10, 6)) 
sns.heatmap(pd.crosstab(df['Maison'], df['Appr√©ciation'], margins=False), annot=True, fmt="d") 
plt.title("Relation entre le type de maison et l'appr√©ciation") 
plt.xlabel('Appr√©ciation') 
plt.ylabel('Type de maison') 
plt.show()
```

```python
# Cr√©er un barplot pour visualiser la distribution des appr√©ciations par type de maison
appreciation = df.groupby('Maison')['Appr√©ciation'].value_counts(normalize=True).unstack().fillna(0) 
appreciation.plot(kind='bar', stacked=True, figsize=(10, 6)) 
plt.title('Distribution des appr√©ciations par type de maison') 
plt.xlabel('Type de maison') 
plt.ylabel("Proportion d'√©tudiants") 
plt.legend(title='Appr√©ciation', bbox_to_anchor=(1.05, 1), loc='upper left') 
plt.show()
```
