```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<h4 id="1">1. Charger le jeu de donnÃ©es "taxis" et afficher les 1eres donnÃ©es<a href="#sommaire">ðŸ‘†</a></h4>

```python
#Dans la variable "df"
df = pd.read_csv("taxis.csv", index_col = 0)
df.head()
```

<h4 id="2">2. Remplacer les noms des colonnes et afficher le resultat<a href="#sommaire">ðŸ‘†</a></h4>

```python
# remplacer les colonnes par :
# "pickup":"depart"
# "dropoff":"arrivee"
# "passengers":"nbpassagers"
# "fare":"tarif"
# "tip":"pourboire"
# "tolls":"peage"
# "color":"couleur"
# "pickup_zone":"zone_depart"
# "dropoff_zone":"zone_arrivee"
# "pickup_borough":"arrondissement_depart"
# "dropoff_borough":"arrondissement_arrivee"
df.rename(columns={"pickup":"depart","dropoff":"arrivee","passengers":"nbpassagers","fare":"tarif","tip":"pourboire","tolls":"peage","color":"couleur",
                  "pickup_zone":"zone_depart","dropoff_zone":"zone_arrivee","pickup_borough":"arrondissement_depart","dropoff_borough":"arrondissement_arrivee"}, inplace=True)
df.head()
```

<h4 id="3">3. Statistique descriptive<a href="#sommaire">ðŸ‘†</a></h4>

```python
df.describe()
```

<h4 id="4">4. Valeurs manquantes<a href="#sommaire">ðŸ‘†</a></h4>

```python
df.isnull().sum()
```

<h4 id="5">5. Distance moyenne parcourue par course<a href="#sommaire">ðŸ‘†</a></h4>

```python
# Calcul de la distance moyenne parcourue par course
print("La distance moyenne parcourue par course est de {} unitÃ©s.".format(df["distance"].mean()))
```

<h4 id="6">6. Diagramme Ã  barres de la distribution des passagers par course<a href="#sommaire">ðŸ‘†</a></h4>

```python
# Nombre de passagers par course
df["nbpassagers"].value_counts()
```

```python
# Diagramme Ã  barres du nombre de passagers par course
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.countplot(df, x ="nbpassagers")
plt.title("Distribution des passagers par course")
plt.xlabel('Nombre de passagers')
plt.ylabel('Nombre de course')
plt.show()
```

<h4 id="7">7. Comparaison du tarif moyen par type de paiement <a href="#sommaire">ðŸ‘†</a></h4>

```python
# Calcul du tarif moyen par type de paiement
df.groupby("payment")[["tarif"]].mean()
```

```python
# Afficher le graphique du tarif moyen par type de paiement (barplot)
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.barplot(df, x = "payment", y = "tarif")
plt.title("Comparaison du tarif moyen par type de paiement")
plt.xlabel('Paiement')
plt.ylabel('Tarif')
plt.show()
```

<h4 id="8">8. Distribution des distances parcourues<a href="#sommaire">ðŸ‘†</a></h4>

```python
# Avec la courbe de densitÃ© (histplot)
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.histplot(df["distance"], kde=True)
plt.title("Distribution des distances parcourues")
plt.xlabel("Distance")
plt.ylabel("Nombres de course")
plt.show()
```

<h4 id="9">9. Distribution des distances parcourues par nombre de passagerd<a href="#sommaire">ðŸ‘†</a></h4>

```python
# BoÃ®te Ã  moustache des distances parcourues par nombre de passagers (boxplot)
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.boxplot(df, y ="distance", x ="nbpassagers")
plt.title("Distribution des distances parcourues par nombres de passagers")
plt.ylabel("Distances parcourues")
plt.xlabel("Nombres de passagers")
plt.show()
```

<h4 id="10">10. Statistiques descriptive des distances parcourues par nombre de passagers<a href="#sommaire">ðŸ‘†</a></h4>

```python
#Statistiques descriptive des distances parcourues par nombre de passagers
df.groupby("nbpassagers")["distance"].describe()
```

<h4 id="11">11. Tableau croisÃ© du nombre de passagers et du type de paiement<a href="#sommaire">ðŸ‘†</a></h4>

```python
# Tableau croisÃ© du nombre de passagers et du type de paiement
pd.crosstab(df['nbpassagers'], df['payment'])
```

```python
# Visualisation (heatmap avec les annotations)
sns.heatmap(pd.crosstab(df['nbpassagers'], df['payment']), annot=True, fmt="d", cmap='Blues')
plt.title('Tableau croisÃ© du nombre de passagers et du type de paiement')
plt.show()
```

<h4 id="12">12. Distribution des pourboires donnÃ©s<a href="#sommaire">ðŸ‘†</a></h4>

```python
# Avec histplot
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.histplot(df["pourboire"], kde=True,  bins = 20)
plt.title("Distribution des pourboires donnÃ©s")
plt.xlabel("ModalitÃ©s")
plt.ylabel("Effectifs")
plt.show()
```

<h4 id="13">13. Boxplot des pourboires en fonction du type de paiement<a href="#sommaire">ðŸ‘†</a></h4>

```python
# Calcul des pourboires en fonction du type de paiement
df.groupby('payment')['pourboire'].mean()
```

```python
# BoÃ®te Ã  moustache des pourboires en foncion du type de paiement
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.boxplot(df, y ="pourboire", x ="payment")
plt.title("BoÃ®te Ã  moustache des pourboires en fonction du type de paiement")
plt.xlabel("Type de paiement")
plt.ylabel("Effectif")
plt.show()
```

<h4 id="14">14. Diagramme en nuage de points de la distance parcourue en fonction du montant payÃ©<a href="#sommaire">ðŸ‘†</a></h4>

```python
# Scatterplot
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.scatterplot(df, x="distance", y="total")
plt.title("Nuage de point de la distance parcourue en fonction du montant payÃ©")
plt.xlabel("Distance parcourue")
plt.ylabel("Montant payÃ©")
plt.show()
```

<h4 id="15">15. CorrÃ©lation entre les variables numÃ©riques<a href="#sommaire">ðŸ‘†</a></h4>

```python
#Avec df.corr
df.drop(df.select_dtypes(include=['object']).columns, axis=1).corr()
```

<h4 id="16">16. Visualisation de la heatmap des corrÃ©lations<a href="#sommaire">ðŸ‘†</a></h4>

```python
sns.heatmap(df.drop(df.select_dtypes(include=['object']).columns, axis=1).corr())
plt.title("Carte de chaleur de la matrice de corrÃ©lation")
plt.show()
```
