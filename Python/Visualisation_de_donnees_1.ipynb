```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<h4 id="1">1. Charger le jeu de donn√©es "diamonds" et afficher les 1eres donn√©es<a href="#sommaire">üëÜ</a></h4>

```python
#Dans la variable "df"
df = pd.read_csv("Datasets/diamonds.csv")
df.head()
```

<h4 id="2">2. Remplacer les noms des colonnes et afficher le resultat<a href="#sommaire">üëÜ</a></h4>

```python
# remplacer les colonnes par :
# "cut":"qualit√©"
# "color":"couleur"
# "clarity":"clart√©"
# "depth":"profondeur"
# "price":"prix"

df.rename(columns = {"cut":"qualit√©", "color":"couleur", "clarity":"clart√©", "depth":"profondeur", "price":"prix"}, inplace = True)
df.head()
```

<h4 id="3">3. Statistique descriptive<a href="#sommaire">üëÜ</a></h4>

```python
df.describe()
```

<h4 id="4">4. Valeurs manquantes<a href="#sommaire">üëÜ</a></h4>

```python
df.isnull().sum()
```

<h4 id="5">5. Prix moyen des diamants<a href="#sommaire">üëÜ</a></h4>

```python
# Calcul du Prix moyen des diamants
df["prix"].mean()
```

<h4 id="6">6. Diagramme √† barres de la distribution des diamants par qualit√©<a href="#sommaire">üëÜ</a></h4>

```python
# Nombre de diamant par qualit√©
df["qualit√©"].value_counts()
```

```python
# Diagramme √† barres de la distribution des diamants par qualit√©
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.countplot(df, x ="qualit√©")
plt.title("Distribution des diamants par qualit√©")
plt.xlabel('Qualit√©')
plt.ylabel('Nombre de diamants')
plt.show()
```

<h4 id="7">7. Comparaison du prix moyen des diamants par qualit√© <a href="#sommaire">üëÜ</a></h4>

```python
# Calcul du prix moyen par qualit√©
df.groupby("qualit√©")[["prix"]].mean()
```

```python
# Afficher le graphique du prix moyen par qualit√© (barplot)
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.barplot(df, x = "qualit√©", y = "prix")
plt.title("Comparaison du prix moyen de diamant par qualit√©")
plt.xlabel('Qualit√©')
plt.ylabel('Prix')
plt.show()
```

<h4 id="8">8. Prix moyen des diamants par couleur<a href="#sommaire">üëÜ</a></h4>

```python
# Avec la courbe de densit√© (histplot)
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.histplot(df, x ="prix", hue ="couleur", kde=True, multiple='stack')
plt.title("Distribution du prix moyen de diamants par couleur")
plt.xlabel("prix de diamants")
plt.ylabel("Effectif")
plt.show()
```

<h4 id="9">9. Visualisation du prix en fonction de la clart√©<a href="#sommaire">üëÜ</a></h4>

```python
# Visualisation du prix en fonction de la clart√©
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.barplot(df, x ="clart√©", y ="prix")
plt.title("Visualisation du prix en fonction de la clart√©")
plt.xlabel('Clart√©')
plt.ylabel('Prix')
plt.show()
```

<h4 id="10">10. Prix moyen par clart√©<a href="#sommaire">üëÜ</a></h4>

```python
# Prix moyen des diamants par clart√©
df.groupby("clart√©")["prix"].mean()
```

<h4 id="11">11. Affichage de la distribution des prix<a href="#sommaire">üëÜ</a></h4>

```python
# Affichage de la distribution des prix
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.histplot(df["prix"], kde=True)
plt.title("Distribution du prix moyen de diamants")
plt.xlabel("prix")
plt.ylabel("Nombres de diamants")
plt.show()
```

<h4 id="12">12. Boxplot du prix en fonction de la qualit√©<a href="#sommaire">üëÜ</a></h4>

```python
# Boxplot des prix en fonction de la qualit√©
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.boxplot(df, y ="prix", x ="qualit√©")
plt.title("Boxplot des prix en fonction de la qualit√©")
plt.ylabel("Prix")
plt.xlabel("Qualit√©")
plt.show()
```

<h4 id="13">13. Statistiques des prix par qualit√© et couleur<a href="#sommaire">üëÜ</a></h4>

```python
# Statistiques des prix par qualit√© et couleur
df.groupby(["qualit√©", "couleur"])["prix"].describe()
```

<h4 id="14">14. Tableau crois√© du prix moyen par qualit√© et couleur et visualisation<a href="#sommaire">üëÜ</a></h4>

```python
# Tableau crois√© du prix moyen par qualit√© et couleur
tabCroise = pd.pivot_table(df, values='prix', index='couleur', columns='qualit√©', aggfunc='mean')
tabCroise
```

```python
# Heatmap du tableau crois√©
sns.heatmap(tabCroise, annot=True, fmt=".2f")
plt.title('Tableau crois√© du prix mmoyen par qualit√© et par couleur')
plt.show()
```

<h4 id="15">15. Distribution du poids des diamants ("carat")<a href="#sommaire">üëÜ</a></h4>

```python
# Distribution du poids des diamants ("carat")
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.histplot(df["carat"], kde=True, bins=20)
plt.title("Distribution du poids des diamants")
plt.xlabel("poids")
plt.ylabel("Nombres de diamants")
plt.show()
```

<h4 id="16">16. Boxplot du poids ("carat") en fonction de la qualit√©<a href="#sommaire">üëÜ</a></h4>

```python
# Boxplot du poids ("carat") en fonction de la qualit√©
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.boxplot(df, y ="carat", x ="qualit√©")
plt.title("Boxplot du poids (carat) en fonction de la qualit√©")
plt.ylabel("Poids")
plt.xlabel("Qualit√©")
plt.show()
```

<h4 id="17">17. Nuage de points du prix en fonction du poids et de la qualit√©<a href="#sommaire">üëÜ</a></h4>

```python
# Nuage de points du prix en fonction du poids et de la qualit√©
sns.set_style("darkgrid", {'grid.color':'grey', 'grid.linestyle': '--'})
sns.scatterplot(df, x="prix", y="carat", hue ="qualit√©")
plt.title("Nuage de points du prix en fonction du poids")
plt.xlabel("Prix")
plt.ylabel("Poids")
plt.show()
```

<h4 id="18">18. Corr√©lation entre les variables num√©riques<a href="#sommaire">üëÜ</a></h4>

```python
# Corr√©lation entre les variables num√©riques
df.drop(df.select_dtypes(include=['object']).columns, axis=1).corr()
```

<h4 id="19">19. Visualisation de la heatmap des corr√©lations<a href="#sommaire">üëÜ</a></h4>

```python
# Heatmap des corr√©lations
sns.heatmap(df.drop(df.select_dtypes(include=['object']).columns, axis=1).corr())
plt.title("Carte de chaleur de la matrice de corr√©lation")
plt.show()
```
